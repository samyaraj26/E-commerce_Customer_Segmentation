```python
import numpy as np
import pandas as pd
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv(r'C:\Users\91986\OneDrive\Desktop\E-comCustomerSegmentation\data\raw_data.csv')
```

```python
print(df.info())
```

```python
df=df.dropna(subset=['CustomerNo'])
```

```python
df=df[~df['TransactionNo'].str.contains('C',na=False)]
```

```python
df = df[(df['Quantity'] > 0) & (df['Price'] > 0)]
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='mixed', dayfirst=False)
df['CustomerNo'] = df['CustomerNo'].astype(int)
```

```python
print(df.info())
```

```python
df['TotalPrice'] = df['Quantity'] * df['Price']
```

```python
file_path = r'C:\Users\91986\OneDrive\Desktop\E-comCustomerSegmentation\data\cleaned_data.csv'
df.to_csv(file_path, index=False)
```
